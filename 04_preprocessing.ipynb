# Data Preprocessing

## Objective

Prepare modelling datasets using leakage-aware preprocessing and train/test separation.

## Outputs

- Training set
- Test set
- Preprocessing pipelines
- Processed feature matrices

### 1. Imports

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings('ignore')
```

### 2. Load the feature-engineered dataset

```python
data = pd.read_csv(
    'feature_engineered_data.csv'
)

print("Dataset Shape:", data.shape)

data.head()
```

```python
data['TARGET'].value_counts(normalize=True) * 100
```

### 3. Create the bureau-history indicator

```python
data['HAS_BUREAU_HISTORY'] = (
    data['BUREAU_LOAN_COUNT'].notna()
).astype(int)
```

```python
data['HAS_BUREAU_HISTORY'].value_counts()
```

```python
data.groupby(
    'HAS_BUREAU_HISTORY'
)['TARGET'].agg(
    ['count', 'mean']
)
```

### 4. Remove identifiers

```python
data = data.drop(
    columns=['SK_ID_CURR']
)
```

### 5. Separate X and y

```python
X = data.drop(
    columns='TARGET'
)

y = data['TARGET']
```

```python
print("X Shape:", X.shape)
print("y Shape:", y.shape)
```

### 6. Train-test split
Because the target is highly imbalanced, use stratify=y.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
# Validate
print("Training Shape:", X_train.shape)
print("Test Shape:", X_test.shape)

print(
    "\nTraining Target Distribution:"
)

print(
    y_train.value_counts(
        normalize=True
    ) * 100
)

print(
    "\nTest Target Distribution:"
)

print(
    y_test.value_counts(
        normalize=True
    ) * 100
)
```

### 7. Identify numerical and categorical features

```python
numerical_features = (
    X_train
    .select_dtypes(
        include=np.number
    )
    .columns
    .tolist()
)

categorical_features = (
    X_train
    .select_dtypes(
        include=['object', 'category']
    )
    .columns
    .tolist()
)

print(
    "Numerical Features:",
    len(numerical_features)
)

print(
    "Categorical Features:",
    len(categorical_features)
)
```

### 8. Find high-missingness columns using training data only

```python
missing_percentage = (
    X_train
    .isnull()
    .mean()
    .mul(100)
    .sort_values(
        ascending=False
    )
)

high_missing_columns = (
    missing_percentage[
        missing_percentage > 60
    ]
    .index
    .tolist()
)

print(
    "Number of columns with >60% missing values:",
    len(high_missing_columns)
)

missing_percentage[
    missing_percentage > 60
]
```

### 9. Drop high-missingness columns

```python
X_train = X_train.drop(
    columns=high_missing_columns
)

X_test = X_test.drop(
    columns=high_missing_columns
)
```

```python
print("Training Shape:", X_train.shape)
print("Test Shape:", X_test.shape)
```

### 10. Recalculate feature types

```python
numerical_features = (
    X_train
    .select_dtypes(
        include=np.number
    )
    .columns
    .tolist()
)

categorical_features = (
    X_train
    .select_dtypes(
        include=['object', 'category']
    )
    .columns
    .tolist()
)

print(
    "Numerical Features:",
    len(numerical_features)
)

print(
    "Categorical Features:",
    len(categorical_features)
)
```

### 11. Inspect missingness after column removal

```python
numerical_missing = (
    X_train[numerical_features]
    .isnull()
    .mean()
    .mul(100)
    .sort_values(
        ascending=False
    )
)

numerical_missing.head(30)
```

```python
categorical_missing = (
    X_train[categorical_features]
    .isnull()
    .mean()
    .mul(100)
    .sort_values(
        ascending=False
    )
)

categorical_missing
```

### 12. Inspect category cardinality

```python
category_cardinality = (
    X_train[categorical_features]
    .nunique()
    .sort_values(
        ascending=False
    )
)

category_cardinality
```

### 13. Define preprocessing pipelines

```python
from sklearn.impute import SimpleImputer

from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

from sklearn.pipeline import Pipeline

from sklearn.compose import ColumnTransformer
```

```python
numerical_pipeline = Pipeline(
    steps=[
        (
            'imputer',
            SimpleImputer(
                strategy='median'
            )
        ),

        (
            'scaler',
            StandardScaler()
        )
    ]
)
```

```python
categorical_pipeline = Pipeline(
    steps=[
        (
            'imputer',
            SimpleImputer(
                strategy='most_frequent'
            )
        ),

        (
            'encoder',
            OneHotEncoder(
                handle_unknown='ignore',
                sparse_output=True
            )
        )
    ]
)
```

```python
# Combined prepreocesor 
preprocessor = ColumnTransformer(
    transformers=[
        (
            'num',
            numerical_pipeline,
            numerical_features
        ),

        (
            'cat',
            categorical_pipeline,
            categorical_features
        )
    ]
)
```

### 14. Fit preprocessing only on training data

```python
X_train_processed = (
    preprocessor.fit_transform(
        X_train
    )
)
```

```python
# Transform test data:
X_test_processed = (
    preprocessor.transform(
        X_test
    )
)
```

### 15. Inspect processed shapes

```python
print(
    "Processed Training Shape:",
    X_train_processed.shape
)

print(
    "Processed Test Shape:",
    X_test_processed.shape
)
```

### 16. Check missing values after preprocessing

```python
if hasattr(
    X_train_processed,
    'data'
):
    print(
        "NaN values in processed training data:",
        np.isnan(
            X_train_processed.data
        ).sum()
    )
else:
    print(
        "NaN values in processed training data:",
        np.isnan(
            X_train_processed
        ).sum()
    )
```

```python
if hasattr(
    X_test_processed,
    'data'
):
    print(
        "NaN values in processed test data:",
        np.isnan(
            X_test_processed.data
        ).sum()
    )
else:
    print(
        "NaN values in processed test data:",
        np.isnan(
            X_test_processed
        ).sum()
    )
```

### 17. Extract transformed feature names

```python
processed_feature_names = (
    preprocessor
    .get_feature_names_out()
)

print(
    "Number of Processed Features:",
    len(processed_feature_names)
)

processed_feature_names[:30]
```

### 18. Save preprocessing objects and split indices

```python
import joblib
import os

os.makedirs(
    'models',
    exist_ok=True
)

joblib.dump(
    preprocessor,
    'models/preprocessor.joblib'
)
```

```python
# Save the dropped columns:
joblib.dump(
    high_missing_columns,
    'models/high_missing_columns.joblib'
)
```

# Preprocessing Summary

1. Created an indicator identifying applicants with matched bureau credit history.

2. Removed the customer identifier because it does not represent a meaningful predictive feature.

3. Split the dataset into training and test sets using stratified sampling to preserve the imbalanced target distribution.

4. Identified and removed features with more than 60% missing values using training data only.

5. Separated numerical and categorical features after column removal.

6. Applied median imputation and standardization to numerical variables.

7. Applied most-frequent imputation and one-hot encoding to categorical variables.

8. Fitted all preprocessing transformations exclusively on training data and applied the learned transformations to the test set to prevent data leakage.

9. Verified that the processed datasets contained no missing values.

10. Saved the fitted preprocessing pipeline for reproducibility and later model evaluation.
